```python
import yfinance as yf
import seaborn as sb
```

```python
companyTickers = ["MSFT","DDOG", "GOOGL", "GTLB", "SNOW", "AAPL","NVDA", "ARM","AAPL", "NET","META"]
etfTickers = ["FNCMX"]
```

```python
# Period can be: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max
data = yf.download(tickers=companyTickers, period="3mo")
```

```python
dayPrice = data.loc[:, "Close"]
sb.heatmap(dayPrice.corr(method="pearson"),cmap="YlGnBu", annot=True)
```
