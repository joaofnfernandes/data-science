```python
import yfinance as yf
import seaborn as sb
import pandas as pd
```

```python
companyTickers = ["MSFT", "DDOG", "GOOGL", "GTLB", "SNOW", "FNCMX", "AAPL", "META", "NVDA"]
# Apple, meta, amazon report today
# NVDIA reports feb 21
```

```python
# Period can be: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max
data = yf.download(tickers=companyTickers, period="15d")
```

```python
def highlight_negative(val):
    color = 'red' if val < 0 else 'black'
    return f'color: {color}'


close = data["Close"].pct_change() * 100
pd.set_option('display.max_rows', None)
trend = close.style.applymap(highlight_negative)
trend
```
