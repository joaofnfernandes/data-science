```python
import yfinance as yf
import seaborn as sb
import pandas as pd
```

```python
companyTickers = ["AAPL", "NVDA", "AMAT", "ASML", "QCOM", "AMD", "AVGO", "TSM", "INTC"]
```

```python
# Period can be: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max
data = yf.download(tickers=companyTickers, period="1mo")
```

```python
def highlight_negative(val):
    color = 'red' if val < 0 else 'black'
    return f'color: {color}'


month = data["Close"].pct_change() * 100
pd.set_option('display.max_rows', None)
trend = month.style.applymap(highlight_negative)
trend
```
